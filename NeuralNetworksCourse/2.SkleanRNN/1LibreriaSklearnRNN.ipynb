```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
```

```python
x = np.array([[x,y,z,w] for x in range(0,2) for y in range(0,2) for z in range(0,2) for w in range(0,2)])
```

```python
y = np.array([a[0] or a[1] or a[2] or a[3] for a in x])
```

```python
clf = MLPClassifier(hidden_layer_sizes=(4,3,2),activation="logistic",
                    learning_rate='adaptive',learning_rate_init=0.7, solver="lbfgs",
                    )
```

```python
clf.fit(x,y)
```

```python
pred = clf.predict(x)
```

```python
print(list(zip(x,y,pred)))
```

```python
print(classification_report(y,pred))
```

```python
modelo = clf.coefs_
```

```python
clf.score(x,y)
```

```python
datos = np.loadtxt("wine.data",delimiter=',')
xI = datos[:,1:]
sal = datos[:,0]
yI = np.array(list(map(lambda a: [0,0,1] if a == 1 else [0,1,0] if a==2 else [1,0,0],sal)))
scaler = StandardScaler()
scaler.fit(xI)
xI = scaler.transform(xI)

x_train,x_test,y_train,y_test = train_test_split(xI, yI, test_size=0.2, random_state=42)
```

```python
clf2 = MLPClassifier(hidden_layer_sizes=(13,10,4),learning_rate="adaptive",learning_rate_init=1.4,solver="lbfgs")
```

```python
clf2.fit(x_train,y_train)
```

```python
clf2.score(x_test,y_test)
```

```python
pred = clf2.predict(x_test)
```

```python
print(classification_report(y_test, pred))
```
