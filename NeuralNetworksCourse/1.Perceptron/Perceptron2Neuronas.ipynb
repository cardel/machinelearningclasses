```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def sigmoide(x):
  if x < -5:
    return 0
  elif x > 5:
    return 1
  else:
    return 1/(1+math.exp(-x))
```

```python
sigmoide_vec = np.vectorize(sigmoide)
```

```python
def neurona(ent,pesos,bias):
  entrada_neta = np.dot(pesos,ent)+bias
  salida = sigmoide_vec(entrada_neta)

  return salida[0]
```

```python
def perceptron(entradas,salidas,factor_aprendizaje,max_iter):
  pesos = 2*np.random.rand(1,entradas.shape[1])-1
  bias = 2*np.random.rand(1)-1
  historico = []
  for it in range(0,max_iter):
    cant_errores = 0
    error_it = 0
    for ent,sal_deseada in zip(entradas,salidas):
      salida_obtenida = neurona(ent,pesos,bias)
      error = sal_deseada - salida_obtenida    
      error_it += abs(error)
      if error != 0:
        cant_errores +=1
        pesos = pesos + factor_aprendizaje*error*ent
        bias = bias+factor_aprendizaje*error*(1)

    historico.append(error_it)
    if cant_errores==0:
      break


  return pesos,bias,historico
```

```python
entradas = np.array([[1,1],[1,0],[0,1],[0,0]])
salidas = np.array([ [ (x[0] and x[1]) , (x[0] or x[1]) ] for x in entradas])
```

```python
#Primera neurona

pesosN1,biasN1,historiaN1 = perceptron(entradas,salidas[:,0],0.9,1000)
print(pesosN1,biasN1)

for ent,sal in zip(entradas,salidas[:,0]):
  print("Entrada ",ent," Salida esperada ",sal," Salida obtenida ",neurona(ent,pesosN1,biasN1))
```

```python
#Segunda neurona

pesosN2,biasN2,historiaN2 = perceptron(entradas,salidas[:,1],0.9,1000)
print(pesosN2,biasN2)

for ent,sal in zip(entradas,salidas[:,1]):
  print("Entrada ",ent," Salida esperada ",sal," Salida obtenida ",neurona(ent,pesosN2,biasN2))
```
