```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
```

```python
modelo = ResNet50()
```

```python
#modelo.summary()
```

```python
!wget https://raw.githubusercontent.com/cardel/repositorios/main/imagen1.jpg
!wget https://raw.githubusercontent.com/cardel/repositorios/main/imagen2.jpg
```

```python
img = image.load_img("imagen2.jpg",target_size=(224,224))
x = image.img_to_array(img)
```

```python
x = np.expand_dims(x,axis=0)
x = preprocess_input(x)
```

```python
x.shape
```

```python
predicion = modelo.predict(x)
print('Predecido:', decode_predictions(predicion, top=3)[0])
```
