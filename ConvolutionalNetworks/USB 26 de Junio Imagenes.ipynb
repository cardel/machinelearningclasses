```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

```python
digitos = datasets.load_digits()
```

```python
imagenes = digitos["images"]
etiquetas = digitos["target"]
```

```python
plt.imshow(imagenes[0])
print(etiquetas[0])
```

```python
plt.imshow(imagenes[1])
print(etiquetas[1])
```

```python
imagenes[2]
```

```python
X = imagenes/255
```

```python
print(X.shape)
```

```python
X2 = X.reshape(1797,64)
```

```python
X2.shape
```

```python
etiquetas
```

```python
X_train, X_test, y_train, y_test = train_test_split(X2,etiquetas,test_size=0.2)
```

```python
modelo = MLPClassifier(hidden_layer_sizes=(64,32),learning_rate_init=0.001)
```

```python
modelo.fit(X_train,y_train)
```

```python
predecidos = modelo.predict(X_test)
```

```python
modelo.score(X_test,y_test)
```

```python
i = 0
pred = predecidos[i]
real = y_test[i]
imagen = X_test[i].reshape(8,8)
plt.imshow(imagen)
print(pred,real)
```

```python
print(classification_report(y_test,predecidos))
```
